```python
import pandas as pd
```

# 1. Call datasets, remove duplicates, NaN.

```python
# Call the clinical dataset and remove duplicates
clinical = pd.read_csv('../Clinical_dataset/clinical_modi_1.csv', sep='\t')
```

```python
# column name change: age_at_index -> age
clinical.rename(columns={'age_at_index':'age'}, inplace=True)
clinical.drop_duplicates(subset='case_submitter_id', inplace=True)
```

```python
# Check missing values in AGE, GENDER, STAGE and remove row with missing.
print('Count of Gender information not available cases:\n', clinical['gender'].isnull().sum(axis=0))
print('\n')
print('Count of STAGE information not available cases:\n', clinical['Stage'].value_counts(dropna=False))
print('\n')
print('Count of AGE information not available cases:\n', clinical['age'].isnull().sum(axis=0))
```

```python
# Remove NaN containing row in 'AGE' & 'STAGE'
clinical = clinical.dropna(subset=['age', 'Stage'], axis=0)
clinical
```

```python
# Call Exposuer dataset and remove duplecates.
exposure = pd.read_csv('../Clinical_dataset/exposure_modi_1.csv', sep='\t')
exposure.drop_duplicates(subset='case_submitter_id',inplace=True)
exposure.head()
```

```python
# Check missing values in PY and remove row with missing.
print('# of PY information not available cases:', exposure['pack_years_smoked'].isnull().sum(axis=0))
exposure = exposure.dropna(subset=['pack_years_smoked'], axis=0)
exposure.head()
```

```python
# Call CCL2 expression dataset (lower 2/3 vs. rest higher)
TREM2_LUAD_L67_H33 = pd.read_csv('./Expression/TREM2_LUAD_L67_H33.csv')
```

# 2. Merge organized datasets and select only necessary columns.

```python
# Sereial merging on 'case_submitter_id'
df_1 = pd.merge(clinical, exposure, on='case_submitter_id')
df_2 = pd.merge(df_1, TREM2_LUAD_L67_H33, left_on='case_submitter_id', right_on='Patient', how='left')
```

```python
# Select only required columns
# 1st check the column names
cols = list(df_2.columns.values.tolist())
cols
```

```python
# then select only necessary columns
df_3 = df_2[['case_submitter_id', 'Patient', 'Status','days_to_death', 'days_to_last_follow_up','Expression', 'Group', 'age', 'gender', 'Stage', 'pack_years_smoked']]
df_3.info()
```

# 3. Change matching variable to numeric form

```python
# change expression to numeric form: THIS PART IS IMPORTANT BECAUSE THIS SETS WHICH IS EXPERIMENT GROUP and WHICH is CONTROL.
print(df_3['Group'].value_counts(dropna=False))

df_3.loc[(df_3.Group == 'Low'), 'Group']=0
df_3.loc[(df_3.Group == 'High'), 'Group']=1
```

```python
# change survival stuatus to numeric form. 
print(df_3['Status'].value_counts(dropna=False))

df_3.loc[(df_3.Status !='Dead'), 'Status']=0
df_3.loc[(df_3.Status == 'Dead'), 'Status']=1
```

```python
# change gender to numeric form.
print(df_3['gender'].value_counts(dropna=False))
df_3.loc[df_3['gender'] == 'male', 'gender'] = 1
df_3.loc[df_3['gender'] == 'female', 'gender'] = 2
```

```python
print(df_3['Status'].value_counts(dropna=False))
print('\n')
print(df_3['Group'].value_counts(dropna=False))
print('\n')
print(df_3['gender'].value_counts(dropna=False))
```

```python
df_3.isnull().sum()
```

```python
# Remove NaN containing row in 'AGE' & 'STAGE'
df_4 = df_3.dropna(subset=['Group', 'Expression','age', 'gender', 'Stage', 'pack_years_smoked'], axis=0)
df_4.isnull().sum()
```

# 4. Save for further analysis

```python
df_4.to_csv('./Results/Merged_TCGA_LUAD_TREM2_L67_H33.csv')
```
